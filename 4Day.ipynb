```python
# 멤버 필드 초기화
class FourCal:
#     first= 0
#     second = 0 
#     result = 0
# 멤버필드를 초기화를 하는 경우에는 멤버필드를 명시하지 않아도 된다.
    def setData(self, num1, num2): # 멤버필드에 값을 주는 것을 초기화라고 한다.
        self.first = num1
        self.second = num2
    def add(self):
        self.result = self.first + self.second
    def sub(self):
        self.result = self.first - self.second

calc = FourCal()
calc.setData(3, 4) # setData함수를 이용해서 초기화 하지 않으면 멤버필드가 정의되지 않아 오류
print(calc.first)
print(calc.second)
calc.add()
print(calc.result)
```

```python
## setData함수를 사용하지 않아도 자동으로 멤버필드가 초기화 되도록 하기.
## 멤버필드 초기화를 객체 생성과 동시에 하자.
class FourCal1:
    # 생성자 : 객체 생성시에 실행되는 함수
    def __init__(self, num1, num2): # 멤버필드를 초기화하기 위한 함수
        self.first = num1
        self.second = num2
    def add(self):
        self.result = self.first + self.second
    def div(self):
        self.result = self.first / self.second
calc1 = FourCal1(32, 45) # __init__함수가 자동으로 실행
## __init__함수가 있는 경우 객체 생성할 때 파라매터값을 전달
#calc2 = FourCal1() # 객체 생성시 클래스명만 적어주면 오류 발생
print(calc1.first)
print(calc1.second)
calc1.add()
print(calc1.result)
```

```python
### 상속
class MoreFourCal(FourCal1): # add(), div()
#     def __init__(self, num1, num2): # 멤버필드를 초기화하기 위한 함수
#         self.first = num1
#         self.second = num2
#     def add(self):
#         self.result = self.first + self.second
#     def div(self):
#         self.result = self.first / self.second
    def mul(self):
        self.result = self.first * self.second
    def sub(self):
        self.result = self.first - self.second
```

```python
class MoreFourCal1(FourCal1): # add(), div()
#     def __init__(self, num1, num2): # 멤버필드를 초기화하기 위한 함수
#         self.first = num1
#         self.second = num2
#     def add(self):
#         self.result = self.first + self.second
#     def div(self):        5      /     0   : 불능
#         self.result = self.first / self.second
    
    def mul(self):
        self.result = self.first * self.second
    def sub(self):
        self.result = self.first - self.second
    # 오버라이딩 :  상속 받은 함수를 다시 정의하는 것 : 재정의
    def div(self): 
        if self.second == 0:
            self.result = 0
        else :
            self.result = self.first / self.second
```

```python
calc6 = MoreFourCal1(5, 0)
calc6.div() # 5 / 0
print(calc6.result)
calc7 = MoreFourCal1(0, 5)
calc7.div()
print(calc7.result)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas
import matplotlib as mpl
import matplotlib.pyplot as plt

source = requests.get("https://www.weather.go.kr/weather/observation/currentweather.jsp")


soup = BeautifulSoup(source.content,"html.parser")
#print(soup.prettify()) 

table = soup.find("table",{'class' : 'table-col'})
# print(table)
data = []

print("*" * 30)
print("오늘 각 지역의 날씨")
print("*" *30 )
#### 크롤링
#print(type(table)) # bs4.element.Tag
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            point = td.find('a').text
            temp = tds[5].text
            humidity = tds[10].text
            #print("{0:<7}{1:<7} {2:<7}".format(point, temp, humidity))
            data.append([point, temp, humidity])
#print(data)
with open('weather.csv', 'w', encoding='utf-8') as f:
    f.write('지역, 온도, 습도\n')
    for i in data:
        f.write("{0},{1},{2}\n".format(i[0], i[1], i[2]))
    
#### 분석
df = pandas.read_csv('weather.csv', index_col='지역', encoding='utf-8')
city_df = df.loc[['서울','인천','대전','대구','광주','부산','울산']]
print(city_df)

font_name = mpl.font_manager.FontProperties(fname='C:\Windows\Fonts\malgun.ttf').get_name()
mpl.rc('font',family=font_name)


ax = city_df.plot(kind='bar',title='날씨',figsize=(12,4),legend=True,fontsize=15)
ax.set_xlabel('도시',fontsize=15)
ax.set_ylabel('기온/습도',fontsize=15)
ax.legend(['기온','습도'],fontsize=15)
 
plt.show()
```

```python
x2 = np.array([1, 2, 3])   # 1차원 배열
grid = np.array([[9, 8, 7], 
                 [6, 5, 4]]) # 2차원 배열
grid1 = np.array([[9, 8, 7], 
                 [1, 2, 3]])
#z = np.concatenate((x2,grid))# 차원이 다르면 안됨
z = np.concatenate((grid,grid1)) # 밑으로 붙는다.
print(z)
z = np.concatenate((grid,grid1), axis = 0) # 밑으로 붙는다.
print(z)
z = np.concatenate((grid,grid1), axis = 1) # 옆으로 붙는다.
print(z)
grid2 = np.array([[9, 8, 7], 
                 [1, 2, 3],
                 [6, 5, 4]])
z = np.concatenate((grid,grid2), axis = 0) # 밑으로 붙는다.
print(z) # 열의 크기 일치해야함
# z = np.concatenate((grid,grid2), axis = 1) # 밑으로 붙는다.
# print(z)
grid3 = np.array([[9, 8], 
                  [1, 2]])
z = np.concatenate((grid,grid3), axis = 1) # 옆으로 붙는다.
print(z) # 행의 크기가 일치 해야한다.

# 차원이 서로 다른 경우
# 1차원 배열과 2차원 배열의 열의 크기가 일치해야 한다.
z = np.vstack([grid, x2]) # 열의 크기가 다르면 합쳐지지 않는다.
print(z)
z = np.vstack([x2 ,grid]) # axis = 0 ,
print(z)

# 
y2 = [[10],[20]] # 열의 길이가 행의 길이와 일치해야 한다.
z = np.hstack([y2, grid]) 
# 행의 크기가 다르면 합쳐지지 않는다. # axis = 1
print(z)
```
